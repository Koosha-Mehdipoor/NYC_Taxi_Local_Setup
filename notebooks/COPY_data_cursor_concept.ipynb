```python
import pandas as pd
import psycopg2
import io

# 1. Create small dataframe
df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["a", "b", "c"]
})

# 2. Create buffer (fake file)
buffer = io.StringIO()

# 3. Write dataframe to buffer as CSV
df.to_csv(buffer, index=False, header=False)

# 4. Go back to beginning
buffer.seek(0)

# 5. Connect to Postgres
conn = psycopg2.connect(
    host="localhost",   # or "postgres" in docker
    database="NYC_Taxi",
    user="koosha",
    password="admin"
)

# 6. Create cursor
cursor = conn.cursor()

# 7. COPY command
cursor.copy_expert(
    "COPY test_copy (id, name) FROM STDIN WITH CSV",
    buffer
)

# 8. Commit
conn.commit()

# 9. Close
cursor.close()
conn.close()
```
